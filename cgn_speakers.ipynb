```python
import pandas as pd
```

```python
recordings = pd.read_excel('./data/recordings.xls')
speakers =  pd.read_excel('./data/speakers.xls')
```

```python
# Explode the speakerIDs column to create multiple rows
exploded_recordings = recordings.assign(SpeakerID=recordings['speakerIDs'].str.split(', ')).explode('SpeakerID')
exploded_recordings
```

```python
# Rename the ID column in the speakers DataFrame
speakers.rename(columns={'ID': 'SpeakerID'}, inplace=True)

# Now you can use the corrected column name to join the datasets
merged_data = pd.merge(exploded_recordings, speakers, on="SpeakerID", how="inner")

# Select the desired columns
selected_columns = merged_data[["recordingID", "SpeakerID", "birthYear", "info", "eduLevel", "occLevel", "firstLang", "homeLang", "sex"]]

# Create a new DataFrame with the selected columns
selected_df = pd.DataFrame(selected_columns)

# Display the DataFrame
selected_df.head(20)
```

```python
# Define mapping dictionaries
edu_mapping = {'edu1': 'hoog', 'edu2': 'midden', 'edu3': 'laag', 'eduX': 'onbekend'}
occ_mapping = {
    'occ1': 'beroep dat een hogere opleiding veronderstelt',
    'occ2': 'beroep dat een middelhoge opleiding veronderstelt',
    'occ3': 'beroep dat een lagere opleiding veronderstelt',
    'occ4': 'ongeschoold beroep',
    'occ5': 'niet-beroepsmatig actief, werkloos',
    'occ6': 'niet-beroepsmatig actief, schoolgaand',
    'occ7': 'niet-beroepsmatig actief, huisvrouw',
    'occ8': 'niet-beroepsmatig actief, arbeidsongeschikt',
    'occ9': 'niet-beroepsmatig actief, anders (b.v met pensioen)',
    'occA': 'werkzaam in het hoger management of bestuur',
    'occB': 'beroep dat een hogere opleiding veronderstelt',
    'occC': 'werkzaam als onderzoeker or docent aan universiteit of instelling voor hoger onderwijs',
    'occD': 'werkzaam in een administratieve functie of een functie in de zorgsector',
    'occE': 'ongeschoold beroep',
    'occF': 'werkzaam als zelfstandige',
    'occG': 'politicus',
    'occH': 'werkzaam in de media (journalist, reporter) of als artiest',
    'occI': 'student, stagiair',
    'occJ': 'niet-beroepsmatig actief',
    'occX': 'onbekend'
}
sex_mapping = {'sex1': 'man', 'sex2': 'vrouw', 'sexX': 'onbekend'}
language_mapping = {'SD': "Standaard Nederlands"}

# Apply mapping to the specified columns
selected_df['eduLevel'] = selected_df['eduLevel'].map(edu_mapping)
selected_df['occLevel'] = selected_df['occLevel'].map(occ_mapping)
selected_df['sex'] = selected_df['sex'].map(sex_mapping)
selected_df['firstLang'] = selected_df['firstLang'].replace(language_mapping)
selected_df['homeLang'] = selected_df['homeLang'].replace(language_mapping)
```
